```python
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1.inset_locator import inset_axes, mark_inset
import numpy as np
```

```python
x = np.linspace(0,2,10000)
y_05 = x**0.5
y_1 = x
y_2 = x**2;
```

```python
fig, ax = plt.subplots(figsize=(6,6), dpi=600)
plt.plot(x, y_05, 'c', linestyle='-', label=r'$\beta=0.5$')
plt.plot(x, y_1, 'm', linestyle='-', label=r'$\beta=1$')
plt.plot(x, y_2, 'y', linestyle='-', label=r'$\beta=2$')

plt.xlabel(r'$S_{ij}$')
plt.ylabel(r'$V_{ij}$', rotation=0, labelpad=10)
plt.xticks([0, 0.5, 1, 1.5, 2])
plt.yticks([0, 0.5, 1, 1.5, 2])
plt.xlim(0,2)
plt.ylim(0,2)
plt.legend()

# 인셋 플롯 추가
ax_inset = ax.inset_axes([0.575, 0.075, 0.38, 0.38])  # zoom을 사용하여 확대 비율 설정
ax_inset.plot(x, y_05, 'c', linestyle='-')
ax_inset.plot(x, y_1, 'm', linestyle='-')
ax_inset.plot(x, y_2, 'y', linestyle='-')

# 인셋 플롯 범위 설정
ax_inset.set_xlim(0, 0.5)
ax_inset.set_ylim(0, 0.5)
ax_inset.set_xticks([0, 0.25, 0.5])
ax_inset.set_yticks([0, 0.25, 0.5])


# 인셋 표시
mark_inset(ax, ax_inset, loc1=2, loc2=4, fc="none", ec="0.5")

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(6,6), dpi=600)
plt.plot(x, y_05, 'c', linestyle='-', label=r'$\beta=0.5$')
plt.plot(x, y_1, 'm', linestyle='-', label=r'$\beta=1$')
plt.plot(x, y_2, 'y', linestyle='-', label=r'$\beta=2$')

plt.xlabel(r'$S_{ij}$')
plt.ylabel(r'$V_{ij}$', rotation=0, labelpad=10)
plt.xticks([0, 0.5, 1, 1.5, 2])
plt.yticks([0, 0.5, 1, 1.5, 2])
plt.xlim(0,2)
plt.ylim(0,2)
plt.legend()

plt.savefig('Fig1.pdf')
```
